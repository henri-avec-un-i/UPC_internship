```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 2D array with random values
data = np.random.rand(10, 10)

# Create a figure and axes
fig, ax = plt.subplots()

# Create the colormap plot
im = ax.imshow(data, cmap='jet')

# Add a colorbar
cbar = ax.figure.colorbar(im, ax=ax)

# Set labels for the colorbar
cbar.ax.set_ylabel('Colorbar Label', rotation=-90, va="bottom")

# Add a title and labels to the axes
ax.set_title('2D Colormap')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Show the plot
plt.show()
```

```python
[Re_min, Re_max, Re_number_of_points] = [0,100,100]
[L_min, L_max, L_number_of_points] = [0,100,100]

# Create the data to plot
Re_list = np.linspace(Re_min, Re_max, Re_number_of_points) #Values of Re for the plot
L_list = np.linspace(L_min, L_max, L_number_of_points) #Values of L (caracteristic lengthscale of the system) for the plot


Re_list_reshaped = np.reshape(Re_list, (Re_number_of_points, 1))
L_list_reshaped = np.reshape(L_list, (L_number_of_points, 1))

#Calculate Kolmogorov lengthscale with eta =~ L/Re^0.75
Re_list_reshaped_raised_to_power = Re_list_reshaped**(-0.75)
eta_array = L_list_reshaped@Re_list_reshaped_raised_to_power.T #Values of Kolmogorov lengthscale (order of magnitude), see formula ref. in Utah University report turblt (add ref.)


print(np.shape(Re_list_reshaped))
print(np.shape(L_list_reshaped.T))
print(np.shape(eta_array))


```
